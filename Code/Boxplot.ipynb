```python
# Metriken + bessere Namen
metrics = {
    "mean_hue_deg": "Farbton",
    "mean_sat": "Sättigung",
    "mean_val": "Helligkeit",
    "colorfulness_hs2003": "Buntheit (Colorfulness)"
}

# Alle Boxplot-Daten sammeln 
records = []
for f in files:
    df = pd.read_csv(f)
    filmname = Path(f).stem.replace("film_boxplot_", "")
    for m in metrics.keys():
        row = df[df["metric"] == m].iloc[0]
        records.append({
            "film": filmname,
            "metric": m,
            "min": row["min"],
            "q1": row["q1"],
            "median": row["median"],
            "q3": row["q3"],
            "max": row["max"]
        })

df_all = pd.DataFrame(records)

# Für jede Metrik ein eigenes Boxplot zeichnen
for m, title in metrics.items():
    df_m = df_all[df_all["metric"] == m].sort_values("film")

    plt.figure(figsize=(10,6))
    for i, row in enumerate(df_m.itertuples()):
        # Whiskers
        plt.plot([i, i], [row.min, row.max], color="black")
        # Box
        plt.plot([i, i], [row.q1, row.q3], color="blue", lw=8)
        # Median
        plt.plot(i, row.median, "ro")

    plt.xticks(range(len(df_m)), df_m["film"], rotation=45, ha="right")
    plt.ylabel(title)
    plt.title(f"Boxplot für {title}")
    plt.tight_layout()
    plt.show()
```
